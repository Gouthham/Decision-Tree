```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
df=pd.read_csv('/content/play_tennis.csv')
df
```

```python
le=LabelEncoder()
df['outlook_n']=le.fit_transform(df['outlook'])
df['temp_n']=le.fit_transform(df['temp'])
df['humidity_n']=le.fit_transform(df['humidity'])
df['wind_n']=le.fit_transform(df['wind'])
df['play_n']=le.fit_transform(df['play'])
```

```python
df
```

```python
df = df.drop(['day', 'outlook', 'temp', 'humidity', 'wind', 'play'], axis=1)

df
```

```python
independent_variable=df.drop('play_n',axis='columns')
dependent_var=df['play_n']
```

```python
model=tree.DecisionTreeClassifier()
#The line model = tree.DecisionTreeClassifier() is used to create an instance of a decision tree classifier model in scikit-learn.
```

```python
model.fit(independent_variable,dependent_var)
#The fit method in ml is used to train a model on a given dataset
#hence the model is trained
```

```python
model.score(independent_variable,dependent_var)
#To predict the accuracy we use model.score
```

```python
model.predict([[1,2,0,1]])
```

```python
plt.figure(figsize=(120,80))
plot_tree(model,filled=True,feature_names=df.columns,class_names=le.classes_)
plt.show()
```
